#### Let's create dummies for the categorical columns

```python
import pandas as pd
df_dummies_estu_prgm_departamento = pd.get_dummies(df_copy['ESTU_PRGM_DEPARTAMENTO'], dtype=int)
df_dummies_fami_tieneinternet = pd.get_dummies(df_copy['FAMI_TIENEINTERNET'], dtype=int)
df_dummies_fami_estratovivienda = pd.get_dummies(df_copy['FAMI_ESTRATOVIVIENDA'], dtype=int)

X2 = pd.concat([df_dummies_fami_tieneinternet, df_dummies_fami_estratovivienda], axis=1)
X4 = pd.concat([df_dummies_estu_prgm_departamento, df_dummies_fami_estratovivienda], axis=1)

y = df_copy['RENDIMIENTO_GLOBAL']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.model_selection import train_test_split

X2tr, X2ts, y2tr, y2ts = train_test_split(X2,y, train_size=0.7)
print(X2tr.shape, X2ts.shape, y2tr.shape, y2ts.shape)

X4tr, X4ts, y4tr, y4ts = train_test_split(X4,y, train_size=0.7)
print(X4tr.shape, X4ts.shape, y4tr.shape, y4ts.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
```
